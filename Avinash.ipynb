```python
data1.boxplot(column=['hospitalizations'],by='oh_region')
title_boxplot = 'Box Plot to detect Outliers'
plt.title( title_boxplot )
plt.suptitle('') # that's what you're after
plt.show()
```

```python
data1.oh_region.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title("Number of records by OH Region")
plt.ylabel("Number of Records")
plt.xlabel("OH Region")
```

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(data1['icu_current_covid_vented'], data1['icu_crci_total_vented'])
ax.set_xlabel("icu_current_covid_vented")
ax.set_ylabel("icu_crci_total_vented")
plt.show()
```
